```python
import requests
from bs4 import BeautifulSoup

# Example of using the libraries to fetch and parse a webpage
url = "https://results.eci.gov.in"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Send HTTP GET request
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    print(soup.prettify())
else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

urls = {
    "Parliamentary Constituencies": "https://results.eci.gov.in/PcResultGenJune2024/index.htm",
    "Andhra Pradesh": "https://results.eci.gov.in/AcResultGenJune2024/index.htm",
    "Odisha": "https://results.eci.gov.in/AcResultGenJune2024/index.htm",
    "Bye Elections": "https://results.eci.gov.in/AcResultByeJune2024/"
}

def scrape_state_data(url):
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=headers)
    data = []
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        tables = soup.find_all('table')
        for table in tables:
            rows = table.find_all('tr')
            for row in rows:
                cols = [ele.text.strip() for ele in row.find_all('td')]
                if cols:
                    data.append(cols)
    else:
        print("Failed to retrieve data from", url)
    return data

all_data = {}
for key, url in urls.items():
    print(f"Scraping data for {key}...")
    data = scrape_state_data(url)
    if data:
        all_data[key] = data

# Visualization using histograms
for key, data in all_data.items():
    df = pd.DataFrame(data, columns=['Constituency', 'Won', 'Leading', 'Total'])
    plt.figure(figsize=(10, 5))
    plt.title(f'Histogram of Data for {key}')
    plt.hist(df['Constituency'], bins=30, alpha=0.7)
    plt.xticks(rotation=45)
    plt.xlabel('Constituency')
    plt.ylabel('Frequency')
    plt.show()
```
